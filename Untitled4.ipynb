Chi-square test (BMI categories vs Outcome)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

df = pd.read_csv('pima_diabetes.csv')

df['BMI_Category'] = pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, 100],
                            labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

contingency_table = pd.crosstab(df['BMI_Category'], df['Outcome'])
print("\nContingency Table:\n", contingency_table)

chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"\nChi-Square Statistic: {chi2:.4f}, p-value: {p:.6f}, dof: {dof}")
print("\nExpected Frequencies:\n", expected)

if p < 0.05:
    print("Result: Significant association between BMI category and diabetes outcome (reject H0).")
else:
    print("Result: No significant association (fail to reject H0).")

contingency_table.plot(kind="bar", stacked=True, figsize=(8,5), colormap="Set2")
plt.title("BMI Category vs Diabetes Outcome")
plt.xlabel("BMI Category")
plt.ylabel("Count")
plt.legend(title="Outcome", labels=["Non-Diabetic (0)", "Diabetic (1)"])
plt.show()
```

One-sample Z-test (BMI vs mean = 25)

```python
import numpy as np
from statsmodels.stats.weightstats import ztest
from scipy.stats import norm

df = pd.read_csv('pima_diabetes.csv')

population_mean = 25
sample = df['BMI']

sample_mean = sample.mean()
sample_std = sample.std(ddof=1)
n = len(sample)

z = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))

print(f"One-sample Z-test:\nZ = {z:.4f}, p = {p:.6f}, sample mean = {sample_mean:.3f}, n = {n}")


plt.figure(figsize=(8,5))
sns.histplot(sample, kde=True, bins=30, color="skyblue")
plt.axvline(population_mean, color="red", linestyle="--", label=f"Population Mean = {population_mean}")
plt.axvline(sample_mean, color="green", linestyle="-", label=f"Sample Mean = {sample_mean:.2f}")
plt.title("BMI Distribution with Population vs Sample Mean")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

Two-proportion Z-test (Obese vs Non-obese)

```python

df['Obese'] = np.where(df['BMI'] >= 30, 1, 0)

obese = df[df['Obese'] == 1]
non_obese = df[df['Obese'] == 0]

p1 = obese['Outcome'].mean()
p2 = non_obese['Outcome'].mean()
n1 = len(obese)
n2 = len(non_obese)

p_pool = (obese['Outcome'].sum() + non_obese['Outcome'].sum()) / (n1 + n2)
z = (p1 - p2) / np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))
p = 2 * (1 - norm.cdf(abs(z)))

print(f"\nTwo-proportion Z-test (Obese vs Non-obese): Z = {z:.4f}, p = {p:.6f}")
print(f"Proportion diabetic in Obese = {p1:.3f} ({obese['Outcome'].sum()}/{n1})")
print(f"Proportion diabetic in Non-obese = {p2:.3f} ({non_obese['Outcome'].sum()}/{n2})")

plt.figure(figsize=(6,5))
sns.barplot(x=["Obese", "Non-Obese"], y=[p1, p2], palette="Set1")
plt.title("Proportion of Diabetics by Obesity Status")
plt.ylabel("Proportion Diabetic")
plt.ylim(0,1)
plt.show()
```

T-tests (BloodPressure & Glucose)

```python
from scipy.stats import ttest_ind

diabetic = df[df['Outcome'] == 1]
non_diabetic = df[df['Outcome'] == 0]

t_bp, p_bp = ttest_ind(non_diabetic['BloodPressure'], diabetic['BloodPressure'], equal_var=False)
print(f"BloodPressure t-test: t = {t_bp:.4f}, p = {p_bp:.6f}")

t_gl, p_gl = ttest_ind(non_diabetic['Glucose'], diabetic['Glucose'], equal_var=False)
print(f"Glucose t-test: t = {t_gl:.4f}, p = {p_gl:.6f}")

fig, axes = plt.subplots(1, 2, figsize=(12,5))
sns.boxplot(x='Outcome', y='BloodPressure', data=df, ax=axes[0], palette="Pastel1")
axes[0].set_title("BloodPressure by Outcome")
axes[0].set_xticklabels(["Non-Diabetic", "Diabetic"])

sns.boxplot(x='Outcome', y='Glucose', data=df, ax=axes[1], palette="Pastel2")
axes[1].set_title("Glucose by Outcome")
axes[1].set_xticklabels(["Non-Diabetic", "Diabetic"])

plt.tight_layout()
plt.show()
```

Pearson Correlation (BMI vs Glucose)

```python
from scipy.stats import pearsonr

r, p_corr = pearsonr(df['BMI'], df['Glucose'])
print(f"\nPearson correlation BMI vs Glucose: r = {r:.4f}, p = {p_corr:.6f}")

plt.figure(figsize=(7,5))
sns.scatterplot(x='BMI', y='Glucose', data=df, alpha=0.6)
sns.regplot(x='BMI', y='Glucose', data=df, scatter=False, color="red")
plt.title("Scatterplot with Regression Line (BMI vs Glucose)")
plt.xlabel("BMI")
plt.ylabel("Glucose")
plt.show()
```
